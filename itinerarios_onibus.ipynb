```python
# import das bibliotecas

import pandas as pd
```

```python
# importa o arquivo diretamente do site do governo

URL = 'https://dados.antt.gov.br/dataset/a8729428-f382-430c-abe5-6e5f85aa9a03/resource/7a387b56-0d45-4326-9056-adf9b070605a/download/horarios.csv'

# abre o arquivo .csv com o separador ';' e encoding 'latin' para conseguir ler os caracteres especiais
df = pd.read_csv(URL, sep=';', na_values=('na','NaN'), encoding='latin')
```

## Razão Social

```python
# conver a coluna horário para datetime

df['time'] =pd.to_datetime(df['horario'])

df.head(2)
```

```python
# imprime os tipo de variável de razão social

print(df['razao_social'].dtypes) 
```

```python
df['razao_social'].value_counts()==2
```

```python
# os nomes que aparecem 2 vezes retornam True
df_2_aux = df['razao_social'].value_counts() == 2

# captura os valores que retornam apenas True e typecastin para lista e organiza em ordem alfabética
lista_2_vezes = df_2_aux[df_2_aux].index.sort_values()

print('Empresas que aparecem apenas duas vezes: \n')
for nome in lista_2_vezes:
     print(nome)
```

```python
# df com as empresas que aparecem apenas 2 vezes

df_2_vezes = df[df['razao_social'].isin(lista_2_vezes)].drop('data_autorizacao',axis=1)
df_2_vezes.head(5)
```

## Intervalo em menos de uma hora

```python
df[['razao_social','horario']].head(5)
```

```python
print(df['horario'].dtype)
```

Os objetos da coluna horário devem ser convertidos para 'datetime'

```python
#tabela dinâmica para relacionar horário e a razao_social
# aggfunc seleciona as horas máximas e mínimas

hora_max = pd.pivot_table(df_2_vezes,index='razao_social',values=['time'],aggfunc='max')
hora_min = pd.pivot_table(df_2_vezes,index='razao_social',values=['time'],aggfunc='min')
```

```python
hora_max.head(5)
```

```python
hora_min.head(5)
```

```python
# calcula o intervalo de tempo entre as séries

intervalo_hora = hora_min.sub(hora_max)
```

```python
print(intervalo_hora.to_string())
```
